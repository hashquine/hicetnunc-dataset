````python
from collections import Counter
from tqdm.notebook import tqdm

cnt = Counter()

images = []

for fpath in tqdm(list(src.config.ipfs0_dir.iterdir())):
    if fpath.stem in {
        'Qmdao8MKmAjkyssQQDLmQRthiXhX1TG4QndZV1kNwoB1SR',
        'QmeipiJCBddE8eHQQHSHFdLPbMtNKxPCMFeeVjsM57q86o',
        'QmR6p19s1LJK121VAwHsN4TcGdW4T9QKu9gj9G5dWtkj34',
        'QmSBc8QuynU7bArUGtjwCRhZUbJyZQArrczKnqM7hZPtfV',
        'QmSVsfwH8es7Ur2eqto9hVpcd2dfWASmEaNxTPpcymuJzg',
    }:
        continue
    config = src.utils.read_json(fpath)
    config_keys = set(config.keys())
    for key in config_keys:
        cnt[key] += 1
    assert config['thumbnailUri'] == 'ipfs://QmNrhZHUaEqxhyLfqoq1mtHSipkWHeT31LNHb1QEbDHgnc'
    assert len(config['formats']) == 1
    assert config['artifactUri'] == config['formats'][0]['uri']
    if config_keys != {
        'name', 'description', 'tags', 'symbol', 'artifactUri', 'creators', 'formats',
        'thumbnailUri', 'decimals', 'isBooleanAmount', 'shouldPreferSymbol'
    }:
        print(fpath, config)
    assert config['artifactUri'].startswith('ipfs://')
    assert config['artifactUri'].count('ipfs://') == 1
    images.append({
        'ipfs': config['artifactUri'].replace('ipfs://', ''),
        'mime': config['formats'][0]['mimeType'],
    })
````

## Convert images

````python
import time
import requests


for image_item in tqdm(images):
    image_ipfs = image_item['ipfs']
    image_mime = image_item['mime']
    raw_image_fpath = src.config.ipfs1_dir / image_ipfs
    if not raw_image_fpath.exists():
        print(image_ipfs, 'not exists')
        continue

    preview_fpath = src.config.previews_dir / (image_ipfs + '.jpeg')
    if preview_fpath.exists():
        continue

    start_time = time.time()

    req = requests.post('http://127.0.0.1:8000/preview/1024x1024', files={
        'file': ('file', raw_image_fpath.read_bytes(), image_mime)
    })
    # print(f'{time.time() - start_time:4.2f}s', end=' ')

    if req.status_code != 200:
        print('Cannot convert ', raw_image_fpath)
        print(req.text)
    else:
        preview_fpath.write_bytes(req.content)
````
